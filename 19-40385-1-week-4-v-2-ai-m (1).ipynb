```python
romanian_map = {
    'Arad': {'Sibiu':140, 'Zerind':75, 'Timisoara':118},
    'Zerind': {'Arad':75, 'Oradea':71},
    'Oradea': {'Zerind':71, 'Sibiu': 151},
    'Sibiu': {'Arad':140, 'Oradea':151, 'Fagaras':99, 'Rimnicu':80},
    'Timisoara': {'Arad': 118, 'Lugoj':111},
    'Lugoj': {'Timisoara':111, 'Mehadia':70},
    'Mehadia': {'Lugoj': 70, 'Drobeta':75},
    'Drobeta': {'Mehadia':75, 'Craiova':120},
    'Craiova': {'Drobeta': 120, 'Rimnicu': 146, 'Pitesti': 138},
    'Rimnicu': {'Sibiu': 80, 'Craiova': 146, 'Pitesti': 97},
    'Fagaras': {'Sibiu':99, 'Bucharest':211},
    'Pitesti': {'Rimnicu': 97, 'Craiova':138, 'Bucharest':101},
    'Bucharest': {'Fagaras':211, 'Pitesti': 101, 'Giurgiu':90, 'Urziceni':85},
    'Giurgiu': {'Bucharest': 90},
    'Urziceni': {'Bucharest': 85, 'Vaslui':142, 'Hirsova':98},
    'Hirsova': {'Urziceni':98, 'Eforie':86},
    'Eforie': {'Hirsova':86},
    'Vaslui': {'Iasi': 92, 'Urziceni':142},
    'Iasi': {'Vaslui':92, 'Neamt': 87},
    'Neamt': {'Iasi':87}
}
```

```python
from queue import Queue


def bfs(startingNode, destinationNode):
    visited = {}#to store explored nodes
    distance = {}
    parent = {}#to store path

    bfs_traversal_output = []#to store the order of traversal
    frontier = Queue()  #Taking a queue named frontier

    for city in romanian_map.keys(): #Loop for initialize variable
        visited[city] = False   #Initializing every city visited as false
        parent[city] = None     #Initializing parent as None
        distance[city] = -1     #Initilizing distance as -1

    startingCity = startingNode   #taking a variable startingCity and initializing it with startingNode
    visited[startingCity] = True  #Here starting city is arad which is setting visited true
    distance[startingCity] = 0
    frontier.put(startingCity)    # arad
    found = False

    while not frontier.empty():
        u = frontier.get()     # in the first iteration it will be arad
        bfs_traversal_output.append(u)  

        for v in romanian_map[u].keys():  #this loop is for traverse all child node
            if not visited[v]:            #if city is not visited is will visit       
                visited[v] = True
                parent[v] = u             #declaring the parent of v node
                print(parent)
                print("\n")
                distance[v] = distance[u] + romanian_map[u][v]                #counting the distance to go from parent node to child node
                frontier.put(v)                                               #putting the childs in frontier queue
                
              
                if v== destinationNode:
                    found =True
                    break
                
        if found: break
            
    g = destinationNode           
    path = []
    while g is not None:
        path.append(g)                # appending bucharest  
        g = parent[g]                 #finding parent
        print(g)
        print("\n")
    
    path.reverse()
    print(path)
    print(distance[destinationNode])
    print(bfs_traversal_output)

bfs('Craiova', 'Arad')
```

# Uniform Cost Search

```python
from queue import PriorityQueue

def ucs(startingNode, destinationNode):
    visited = {}  # to store explored note
    distance = {} # to store the distance
    parent = {}   #to keep track of the parents

    ucs_traversal_output = []
    pq = PriorityQueue()        #taking priority queue which will be prioritize by the distance of the city

    for city in romanian_map.keys():          #initializing every visited as false , parent as none distance as -1 
        visited[city] = False
        parent[city] = None
        distance[city] = -1

    startingCity = startingNode                #taking a variable startingCity and initializing it with startingNode
    visited[startingCity] = True               #Here starting city is arad which is setting visited true
    distance[startingCity] = 0
    pq.put((0,startingCity))                   #arad 

    while not pq.empty():
        u = pq.get()[1]                        # in the first iteration it will be arad   
        ucs_traversal_output.append(u)
        if u== destinationNode:               #if we find the destination we will break the loop
            break
        visited[u] = True
        for v in romanian_map[u].keys():
            if not visited[v]:                
                parent[v] = u                 #exploring  the parent of the node
                distance[v] = distance[u] + romanian_map[u][v]             #counting distance
                pq.put((distance[v],v))             #putting in queue based on priority
               

    
    g = destinationNode
    path = []
    while g is not None:
        path.append(g)
        g = parent[g]                     #finding the parent 
        print(g)
        print("\n")
        
    path.reverse()
    print(path)
    print(distance[destinationNode])
    print(ucs_traversal_output)


ucs('Craiova', 'Arad')
```
