# Problem: Maximie f(x) = x^2 when x= 1 to 31

 # Step 1: Initialize Population

```python
import random
def initPopulation(n,b):
    p={}
    for i in range(n):
        s=[]
        for j in range(b):
            s.append(random.choice([0,1]))    
        p[i]=s
    return p      
```

```python
population=initPopulation(4,5)
population
```

 # # Step 2: Parent Selection

**Fitness**

```python
def getFitness(p):
    f=[]
    for k in p.keys():
        d = 0
        b = len(p[k])
        for i in range(b-1,-1,-1):
            d=d+p[k][i]*2**(4-i)
        f.append(d**2)
    return f
```

```python
fitness = getFitness(population)
fitness
```

Probability

```python
total_fitness = sum(fitness)
prob = [round(f/total_fitness,3) for f in fitness]
prob
```

```python
def selectParent(n,probabilities):
    parents = []
    for i in range(n):
        s = 0
        r =random.choice(range(1,1001,1))/1000 
        for j in range(len(probabilities)):
            s = s + probabilities[j]   
            if(s>r):
                parents.append(j)
                break   
    return parents
parents=selectParent(4,prob)
```

# Step 4: Survivor (will go to population)

```python
def crossover(parent_1,parent_2,point):
    p1,p2 = list(parent_1),list(parent_2)   
    for i in range(point,len(p1)):
        p1[i],p2[i] = p2[i],p1[i]         
    p1,p2 = ''.join(p1),''.join(p2)     
    return p1,p2
```

```python
parent_1 = '10011'
parent_2 = '01001'      
print("\n")
print('Parent(1):',parent_1)               
print('Parent(2):',parent_2)
print("\n")
point = random.choice(range(1,5))      
print("\n")
print('crossover point:',point)   
offSpring_1,offSpring_2 = crossover(parent_1,parent_2,point)
print("\n")
print('Offspring(1):',offSpring_1)     
print('Offspring(2):',offSpring_2)
```
