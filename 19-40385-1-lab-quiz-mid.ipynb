```python
table={('B','Clean'):'go A',('B','Dirty'):'suck', ('C','Clean'):'go B',('C','Dirty'):'suck',('A','Clean'):'go C',('A','Dirty'):'suck'}
percepts=[]    
def table_driven_agent(percept):
    print(percept)
    percepts.append(percept)
    action = check(percept,table)
    return action
def check(percept,table):
    for key in table:
        if key == percept:
            return table[key]
import random
Location=random.choice(['A','B','C'])
Condition=random.choice(['Clean','Dirty'])
while True:
    action=table_driven_agent((Location,Condition))
    print(action)
    n=input()
    if(n=='end'): 
        break
    elif action=='go B':
        Location='B'
        Condition=random.choice(['Clean','Dirty'])
    elif action=='go C':
        Location='C'
        Condition=random.choice(['Clean','Dirty'])
    elif action=='go A':
        Location='A'
        Condition=random.choice(['Clean','Dirty'])
    else:
        Condition='Clean'   
```
