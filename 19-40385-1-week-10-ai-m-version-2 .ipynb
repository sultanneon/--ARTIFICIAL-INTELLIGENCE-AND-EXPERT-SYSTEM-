# N Queen Attack Detection

```python
import random
```

```python
def board(N):
    N = 4
    r = []
    c = []
    random.shuffle(r)
    random.shuffle(c)
    for i in range(1, N+1):
        r.append(random.choice(range(1,N+1)))
        c.append(random.choice(range(1,N+1)))
        print('Position Q{}:{},{}'.format(i,r[-1],c[-1]))
    return [r,c]
```

```python
# row attack
def rowa(r,c,N):
    for i in range(0,N):
        for j in range(i+1,N):
            if(r[i] ==r[j]):
                print('Row Attack:Row:{} Column:{} and {}'.format(r[i],c[i],c[j]))   
                return 0
    
    return 1    
    
```

```python
#column attack
def cola(r,c,N):
    for i in range(0,N):
        for j in range(i+1,N):
            if(c[i] ==c[j]):
                print('Column Attack: Row:{} and {} Column:{}'.format(r[i],r[j],c[j]))      
                return 0
    
    return 1
```

```python
# diagonal attack
def diaa(r,c,N):
    for i in range(0,N):
        for j in range(i+1,N):
            if(abs(c[i]-c[j]) ==abs(r[i]-r[j])):
                print('Diagonal Attack: Position:{},{} and {},{}'.format(c[i],r[i],c[j],r[j]))  
                return 0
    
    return 1
```

```python
# implement a method to print a state is valid or not
# input r and c (two lists)
N=4;l=board(N)
r=l[0]
c=l[1]
print(c)
print(r)
col=cola(r,c,N)
row=rowa(r,c,N)
dia=diaa(r,c,N)
if(col==0 or row==0 or dia==0):
    print('Here state is not valid') 
else:
    print('Here state is valid')
    
```
