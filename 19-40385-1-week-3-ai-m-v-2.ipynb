**Table driven agent**

```python
table={('Present','Day'):'OFF',('Present','Night'):'ON',('Absence','Day'):'OFF',('Absence','Night'):'OFF'}
percepts=[]    
def table_driven_agent(percept):
    print(percept)
    percepts.append(percept)
    action = lookup(percept,table)
    return action
def lookup(percept,table):
    for key in table:
        if key == percept:
            return table[key]
import random
Available = random.choice(['Present','Absence'])
Condition = random.choice(['Day','Night'])
while True:
    action= table_driven_agent((Available,Condition))
    print(action)
    cmd = input()
    if(cmd == 'finish'): 
        break
    if action == 'ON':
        Available = 'Absence'
        Condition = random.choice(['Day','Night'])
    elif action== 'OFF':
        Available = 'Present'
        Condition = random.choice(['Day','Night'])
    else:
        Condition = 'Day'  
    
```

**Simple reflex agent**

```python
def simple_reflex_agent(percept):
    print(percept)
    if percept[0]=='Present':
        if percept[1]=='Night':
            action = 'ON'
        else :
            action = 'OFF'
    else: 
        action = 'OFF'
    return action
import random
Available = random.choice(['Present','Absence'])
Condition = random.choice(['Day','Night'])
while True:
    action= simple_reflex_agent((Available,Condition))
    print(action)
    cmd = input()
    if(cmd == 'finish'): 
        break
    if action == 'ON':
        Available = 'Absence'
        Condition = random.choice(['Day','Night'])
    elif action== 'OFF':
        Available = 'Present'
        Condition = random.choice(['Day','Night'])
    else:
        Condition = 'Day'  
```

**Model based agent**

```python
model = {'Day':'Unknown', 'Night':'Unknown'}
state = 'Bad'
action = 'Unknown'

def update_mode(percept, model):
    global state, action
    if action=='ON':
        model[percept[0]]='Visible'
    if model['Day']=='Visible' and model['Night']=='Visible':
        state = 'Good'
    else:
        state = 'Bad'
    return state
        

def model_based_reflex_agent(percept):
    global state,action,model
    model[percept[0]]=percept[1]
    state = update_mode(percept, model)
    print('Perception: '+str(percept)) 
    print('Model: '+str(model))   
    print('State: '+str(state))  
    if state == 'Good':
        action = 'Nothing to do'
    elif percept[1]=='Not visible':
        action = 'ON'
    elif percept[0]=='Day':
        action = 'OFF'
    elif percept[0]=='Night':
        action = 'ON'
    return action


import random
mode = random.choice(['Day','Night'])
Condition= random.choice(['Visible','Not visible'])

while True:
    action = model_based_reflex_agent((mode,Condition))
    print('Action: '+ str(action))
    if action == 'ON':
        mode = 'Day'
        Condition = random.choice(['Visible','Not visible'])
    elif action == 'OFF':
        mode = 'Night'
        Condition = random.choice(['Visible','Not visible'])
    elif action == 'Nothing to do':
        break
        

    
    
```
